```python
# Dependencies
import json
import requests
import pprint
from NPS_config import api_key
```

```python
#config information from Amenities API branch
url1 = 'https://developer.nps.gov/api/v1/amenities?'
# city = ??

#build query url
query_url1 = url1 + "&api_key=" + api_key 
# + city
```

```python
# Get NPS amenity data
nps_response = requests.get(query_url1)
nps_amenity_json = nps_response.json()

# Get the amenitites from the response
print(f"The NPS API responded with the following amenity information: {nps_amenity_json}.")
```

```python
# Get the data list from the JSON response
amenities_data = nps_amenity_json.get('data', [])

for amenity_info in amenities_data:
    amenity_id = amenity_info.get('id', 'N/A')
    amenity_name = amenity_info.get('name', 'N/A')
    amenity_categories = ', '.join(amenity_info.get('categories', []))

    print(f'ID: {amenity_id}')
    print(f'Name: {amenity_name}')
    print(f'Categories: {amenity_categories}')
    print('---------------------------')
```
