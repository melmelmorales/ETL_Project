```python
# Remove "[" and "]" from the "tags" column
NPS_ThingsToDo_df['tags'] = NPS_ThingsToDo_df['tags'].str.strip('[]').str.replace("'", "")

# Split the "tags" column into multiple columns dynamically
tags_split = NPS_ThingsToDo_df['tags'].str.split(',', expand=True)

# Rename the columns to tag1, tag2, tag3, and so forth
tags_split.columns = [f"tag{i+1}" for i in range(tags_split.shape[1])]

# Concatenate the split columns with the original DataFrame
NPS_ThingsToDo_df = pd.concat([NPS_ThingsToDo_df, tags_split], axis=1)

# Display the updated DataFrame
NPS_ThingsToDo_df.head(5)
```

```python
# Remove "[" and "]" from the "season" column
NPS_ThingsToDo_df['season'] = NPS_ThingsToDo_df['season'].str.strip('[]').str.replace("'", "")

# Split the "season" column into multiple columns dynamically
season_split = NPS_ThingsToDo_df['season'].str.split(',', expand=True)

# Rename the columns to season1, season2, season3, and so forth
season_split.columns = [f"season{i+1}" for i in range(season_split.shape[1])]

# Concatenate the split columns with the original DataFrame
NPS_ThingsToDo_df = pd.concat([NPS_ThingsToDo_df, season_split], axis=1)

# Display the updated DataFrame
NPS_ThingsToDo_df.head(5)
```

```python
# Remove "[" and "]" from the "activities" column
NPS_ThingsToDo_df['activities'] = NPS_ThingsToDo_df['activities'].str.strip('[]').str.replace("'", "")

NPS_ThingsToDo_df.head(5)
```

```python
# Drop the original "tags" and "seasons" columns
NPS_ThingsToDo_df = NPS_ThingsToDo_df.drop(columns=['tags', 'season'])

NPS_ThingsToDo_df.head(5)
```

```python
# Rename columns to align with other DataFrames
NPS_ThingsToDo_df.rename(columns={
    "parkCode": "park_code", 
    "fullName": "park_name", 
    "designation": "park_designation",
    "isReservationRequired": "reservation_required",
    "arePetsPermitted": "pets_permitted",
    "doFeesApply": "do_fees_apply",
    "location": "location",
    "duration": "duration",
    "states": "park_states"
}, inplace=True)

print("\nTransformed NPS_ThingsToDo_df")
NPS_ThingsToDo_df.head(5)
```

```python
# Drop the duration column in NPS_ThingsToDo_df
NPS_ThingsToDo_df = NPS_ThingsToDo_df.drop(columns=['duration'])
```

### Step 3: Edit/merge data to create appropriate datasets for use in analysis

```python
# Define a function to handle merging and combining columns
def merge_dataframes(left_df, right_df, on, suffixes=('_x', '_y')):
    merged_df = left_df.merge(right_df, on=on, how="left", suffixes=suffixes)
    for col in left_df.columns:
        if col + suffixes[0] in merged_df.columns and col + suffixes[1] in merged_df.columns:
            merged_df[col] = merged_df[col + suffixes[0]].combine_first(merged_df[col + suffixes[1]])
            merged_df = merged_df.drop(columns=[col + suffixes[0], col + suffixes[1]])
    return merged_df
```

```python
# Merge DataFrames and combine columns for Amenities by Place
amenities_by_place_df = merge_dataframes(NPS_ParkCodes_df, NPS_AmenitiesByPlace_df, on="park_code")
amenities_by_place_df = merge_dataframes(amenities_by_place_df, NPS_Parks_df, on="park_code")
amenities_by_place_df = merge_dataframes(amenities_by_place_df, NPS_Weather_df, on="park_code")
```

```python
# Display info and head of the merged DataFrames
print("\nAmenities by Place DataFrame info")
print(amenities_by_place_df.info())
amenities_by_place_df.head(5)
```

```python
# Merge DataFrames and combine columns for Activities by Park
activities_by_park_df = merge_dataframes(NPS_ParkCodes_df, NPS_ActivitiesByPark_df, on="park_code")
activities_by_park_df = merge_dataframes(activities_by_park_df, NPS_Parks_df, on="park_code")
activities_by_park_df = merge_dataframes(activities_by_park_df, NPS_Weather_df, on="park_code")
```

```python
print("\nActivities by Park DataFrame info")
print(activities_by_park_df.info())
activities_by_park_df.head(5)
```

```python
# Merge the Amenities and Activities DataFrames
amenities_activities_df = pd.concat([amenities_by_place_df, activities_by_park_df], ignore_index=True)
```

```python
# Display info and head of the merged DataFrame
print("\nAmenities and Activities DataFrame info")
print(amenities_activities_df.info())
amenities_activities_df.head(5)
```

```python
# Merge DataFrames and combine columns for Entrance Fees
entrance_fees_df = merge_dataframes(NPS_ParkCodes_df, NPS_EntranceFees_df, on="park_code")
entrance_fees_df = merge_dataframes(entrance_fees_df, NPS_Parks_df, on="park_code")
entrance_fees_df = merge_dataframes(entrance_fees_df, NPS_Weather_df, on="park_code")
```

```python
print("\nEntrance Fees DataFrame info")
print(entrance_fees_df.info())
entrance_fees_df.head(5)
```

```python
# Merge DataFrames and combine columns for Things To Do
things_to_do_df = merge_dataframes(NPS_ParkCodes_df, NPS_ThingsToDo_df, on="park_code")
things_to_do_df = merge_dataframes(things_to_do_df, NPS_Parks_df, on="park_code")
things_to_do_df = merge_dataframes(things_to_do_df, NPS_Weather_df, on="park_code")
```

```python
print("\nThings to Do DataFrame info")
print(things_to_do_df.info())
things_to_do_df.head(5)
```

```python
# Define the directory path for saving the CSV files
directory_path = Path("../Transform/transformed_nps_csv_files")

# Create the directory if it doesn't exist
directory_path.mkdir(parents=True, exist_ok=True)

# Save the merged DataFrames to CSV files in the specified directory
amenities_activities_df.to_csv(directory_path / "nps_amenities_activities.csv", index=False)
entrance_fees_df.to_csv(directory_path / "nps_entrance_fees.csv", index=False)
things_to_do_df.to_csv(directory_path / "nps_things_to_do.csv", index=False)
```
